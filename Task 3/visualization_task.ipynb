```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
print("Loading data...")

```

```python
sns.get_dataset_names()
```

```python
taxis_df=sns.load_dataset("exercise")
taxis_df.head()
```

```python
sns.lineplot(data=taxis_df,x="time",y="pulse",hue="kind")
plt.title("Line plot of Time vs Pulse (Seaborn)")
plt.show()
```

```python
sns.barplot(data=taxis_df,x="kind",y="pulse")
plt.title("Bar plot of Kind vs Pulse (Seaborn)")
plt.show()
```

```python
diamonds_df=sns.load_dataset("diamonds")
diamonds_df.head()
```

```python
sns.scatterplot(data=diamonds_df,x="carat",y="price",hue="cut")
plt.title("Scatter plot of Carat vs Price(Seaborn)")
plt.show()
```

```python
cut_colors = {'Fair':'red', 'Good':'orange', 'Very Good':'green', 'Premium':'blue', 'Ideal':'purple'}
plt.figure(figsize=(8, 5))
for cut, group in diamonds_df.groupby("cut"):
    plt.scatter(group["carat"], group["price"], s=10, alpha=0.5, label=cut, c=cut_colors.get(cut, "gray"))
plt.xlabel("carat")
plt.ylabel("price")
plt.title("Scatter plot of Carat vs Price (matplotlib)")
plt.legend(title="cut")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=diamonds_df,x="cut",y="price")
plt.title("Average Price by Cut (Seaborn)")
plt.show()
```

```python
sns.boxplot(data=diamonds_df,x="cut",y="price")
plt.title("Price Distribution by Cut (Seaborn)")
plt.show()
```
